```python
import pandas as pd
dataset_df = pd.read_csv('https://raw.githubusercontent.com/m37335/kanagawa-exam/master/data/textbook_sentence_maskedword_df-check.csv')
```

```python
dataset_df
```

```python
sentences = dataset_df.sentence.values.tolist()
masked_words = dataset_df.masked_Word.values.tolist()

print(sentences[0:2])
print(sentences[0].replace('band', '[MASK]'))
```

```python
sentence_list = []
for sentence, masked_word in zip(sentences, masked_words):
  sentence_list.append(sentence.replace(masked_word, '[MASK]'))
```

```python
sentence_list[0:9]
```

```python
df = pd.DataFrame(sentence_list,
                  columns=['sentence'])
```

```python
df['masked_word'] = masked_words
```

```python
df
```

```python
# CSVで出力し、ダウンロード
from google.colab import files
filename =  "textbook_sentence_maskedword_df.csv"
df.to_csv(filename, encoding = 'utf-8-sig') 
files.download(filename)
```
